```python
def match_name1(params):
    '''
    Matches scientific names using the GBIF Species API.
    Returns GBIF JSON.
    '''
    url = 'http://api.gbif.org/v1/species/match'
    r = requests.get(url, params=params)
    return(r.json())


def match_name(name):
    '''
    Matches scientific names using the GBIF Species API.
    Returns GBIF JSON.
    '''
    url = 'http://api.gbif.org/v1/species/match'
    params = {'name': name, 'verbose': False}
    r = requests.get(url, params=params)
    return(r.json())
```

```python
# Build a list containing JSON responses from GBIF name matches
mylist = []
for index, row in df.iterrows():
    print(index, row.scientificName) # Show progress
    mydict = {'target': row.scientificName}
    params = {'name': row.scientificName, 'kingdom': row.kingdom, 'rank': row.taxonRank}
    mydict.update(match_name1(params))
    mylist.append(mydict)
    
# Load list into a pandas data frame
df_results = pd.DataFrame(mylist) 

# Convert key columns from float to integer.
# Note that missing values (NA) are flagged with -1.
for colname in df_results.columns:
    if 'Key' in colname:
        print(colname)
        df_results[colname] = df_results[colname].fillna(-1.0).astype(int)
df_results
```
